```python
from tkinter import *
import joblib
window=Tk()
window.title('flood')
window.geometry('400x600')
l0=Label(window,text='jan')
l0.grid(row=0,column=1)
e0=Entry(window)
e0.grid(row=0,column=2)


l1=Label(window,text='frb')
l1.grid(row=1,column=1)
e1=Entry(window)
e1.grid(row=1,column=2)


l2=Label(window,text='mar')
l2.grid(row=2,column=1)
e2=Entry(window)
e2.grid(row=2,column=2)


l3=Label(window,text='apr')
l3.grid(row=3,column=1)
e3=Entry(window)
e3.grid(row=3,column=2)


l4=Label(window,text='may')
l4.grid(row=4,column=1)
e4=Entry(window)
e4.grid(row=4,column=2)


l5=Label(window,text='jun')
l5.grid(row=5,column=1)
e5=Entry(window)
e5.grid(row=5,column=2)


l6=Label(window,text='jul')
l6.grid(row=6,column=1)
e6=Entry(window)
e6.grid(row=6,column=2)


l7=Label(window,text='aug')
l7.grid(row=7,column=1)
e7=Entry(window)
e7.grid(row=7,column=2)


l8=Label(window,text='sep')
l8.grid(row=8,column=1)
e8=Entry(window)
e8.grid(row=8,column=2)


l9=Label(window,text='octo')
l9.grid(row=9,column=1)
e9=Entry(window)
e9.grid(row=9,column=2)


l10=Label(window,text='nov')
l10.grid(row=10,column=1)
e10=Entry(window)
e10.grid(row=10,column=2)


l11=Label(window,text='dec')
l11.grid(row=11,column=1)
e11=Entry(window)
e11.grid(row=11,column=2)

def clicked():
    print('button clicked')
    jan=e0.get()
    frb=e1.get()
    mar=e2.get()
    apr=e3.get()
    may=e4.get()
    jun=e5.get()
    jul=e6.get()
    aug=e7.get()
    sep=e8.get()
    octo=e9.get()
    nov=e10.get()
    dec=e11.get()
    print(jan,frb,mar,apr,may,jun,jul,aug,sep,octo,nov,dec)
    model=joblib.load('knn.joblib')
    p=model.predict([[jan,frb,mar,apr,may,jun,jul,aug,sep,octo,nov,dec]])
    print(p)

b=Button(window,text='predict',command=clicked)
b.grid(row=12,column=1)
window.mainloop()
```
